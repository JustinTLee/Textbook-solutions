## Problem:

**[Purpose: For you to see the influence of the prior in each successive flip, and for you to see another demonstration that the posterior is invariant under re-orderings of the data.]**

For this exercise, use the R function explained in Section 6.6 (`BernBeta.R`). (Don’t forget to source the function before calling it.) Notice that the function returns the posterior beta values each time it is called, so you can use the returned values as the prior values for the next function call.

(A) Start with a prior distribution that expresses some uncertainty that a coin is fair: beta(θ|4, 4). Flip the coin once; suppose we get a head. What is the posterior distribution?

(B) Use the posterior from the previous flip as the prior for the next flip. Suppose
we flip again and get a head. Now what is the new posterior? (Hint: If you type `post = BernBeta( c(4,4) , c(1) )` for the first part, then you can type `post = BernBeta( post , c(1) )` for the next part.)

(C) Using that posterior as the prior for the next flip, flip a third time and get a tail. Now what is the new posterior? (Hint: Type `post = BernBeta( post , c(0) )`.)

(D) Do the same three updates but in the order T, H, H instead of H, H, T. Is the final posterior distribution the same for both orderings of the flip results?

## Solution:

Source `BernBeta.R`:

```r
source("../DBDA2Eprograms/BernBeta.R")
```

#### A.

We start with a $\text{beta}(\theta \, | \, a = 4, b = 4)$ prior. One coin is flipped with the outcome 'H'. The prior, likelihood, and posterior look like this:

```r
flip1 = BernBeta(c(4, 4), c(1))
```

The posterior looks like the prior, except shifted towards a higher value of $\theta$.

#### B.

The posterior from part A will be used as the prior for part B. Another coin flip generates a head:

```r
flip2 = BernBeta(flip1, c(1))
```

The posterior distribution has again shifted towards higher values of $\theta$.

#### C.

The posterior from part B will be used as the prior for part C. Another coin flip generates a tails:

```r
flip3 = BernBeta(flip2, c(0))
```

The posterior distribution has shifted towards lower values of $\theta$.

#### D.

We will update the distribution with T, H, H:

```r
flip4 = BernBeta(flip3, c(0))
flip5 = BernBeta(flip4, c(1))
flip6 = BernBeta(flip5, c(1))
```

This brings us back to the posterior with almost the same shape as the initial update in part A.
