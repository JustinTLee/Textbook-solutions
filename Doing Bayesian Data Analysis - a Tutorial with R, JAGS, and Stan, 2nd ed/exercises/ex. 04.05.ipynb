## Problem:

**[Purpose: To have you use a normal curve to describe beliefs. It’s also handy to know the area under the normal curve between μ and σ .]**

(A) Adapt the code from IntegralOfDensity.R to determine (approximately) the probability mass under the normal curve from x = μ − σ to x = μ + σ . Comment your code. Hint: Just change xlow and xhigh appropriately, and change the text location so that the area still appears within the plot. <br/>
(B) Now use the normal curve to describe the following belief. Suppose you believe that women’s heights follow a bell-shaped distribution, centered at 162 cm with about two-thirds of all women having heights between 147 and 177 cm. What should be the μ and σ parameter values?

## Solution:

#### A.

```r
source("../DBDA2Eprograms/DBDA2E-utilities.R")

meanval = 0.0            # mu value (unchanged from original code)
sdval = 0.2              # sigma value (unchanged from original code)
xlow = meanval - sdval   # mu - sigma
xhigh = meanval + sdval  # mu + sigma
dx = sdval/40            # step size

# Specify comb of points along the x axis:
x = seq(from = xlow, to = xhigh, by = dx)

# Compute y values, i.e., probability density at each value of x:
y = (1/(sdval*sqrt(2*pi)))*exp(-0.5 * ((x - meanval) / sdval)^2)

plot(x, y, type = "h", lwd = 1,
     main = "Probability density p(x)",
     xlab = "x",
     ylab = "p(x)")

area = sum(dx * y)

text(max(x), max(y),
     bquote(paste(sum(, x,) , " ", Delta, "x p(x) = ", .(signif(area, 3)))),
     adj = c(1.0, 0.5), cex = 1.5)
```

The probability contained within the first standard deviation is approximately 66%.

#### B.

```r
source("../DBDA2Eprograms/DBDA2E-utilities.R")

meanval = 162            # mu value (unchanged from original code)
sdval = 15               # sigma value (unchanged from original code)
xlow = meanval - sdval   # mu - sigma
xhigh = meanval + sdval  # mu + sigma
dx = sdval/40            # step size

# Specify comb of points along the x axis:
x = seq(from = xlow, to = xhigh, by = dx)

# Compute y values, i.e., probability density at each value of x:
y = (1/(sdval*sqrt(2*pi)))*exp(-0.5 * ((x - meanval) / sdval)^2)

plot(x, y, type = "h", lwd = 1,
     main = "Probability density p(x)",
     xlab = "height (cm)",
     ylab = "p(height)")

area = sum(dx * y)

text(max(x), max(y),
     bquote(paste(sum(, x,) , " ", Delta, "x p(x) = ", .(signif(area, 3)))),
     adj = c(1.0, 0.5), cex = 1.5)
```

Around 2/3 of women fall within the 147 to 177 cm height range.
