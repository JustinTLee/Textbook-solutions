## Problem:

**[Purpose: Getting experience with the details of the command syntax within R.]**

Adapt the program SimpleGraph.R so that it plots a cubic function (y = x3 ) over the interval x ∈ [−3, +3]. Save the graph in a file format of your choice. Include a properly commented listing of your code, along with the resulting graph.

## Solution:

```r
source("../DBDA2Eprograms/DBDA2E-utilities.R")      # read defn. of openGraph, saveGraph
openGraph(width = 3, height = 4)                    # open a graphics window
x = seq(from = -3, to = 3, by = 0.1)                # Specify vector of x values.
y = x^3                                             # Specify corresponding y values.
plot(x, y, col = "black", type = "l",               # Plot the x,y points as a black line.
     main = "x^3",
     xlab = "x",
     ylab = "x^3")
saveGraph(file = "../images/cubic", type = "png")   # save graph as PNG
```
