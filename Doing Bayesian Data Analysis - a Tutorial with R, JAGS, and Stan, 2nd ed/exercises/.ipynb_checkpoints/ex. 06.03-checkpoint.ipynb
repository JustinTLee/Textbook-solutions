# Exercise 06.02

## Problem:

**[Purpose: Connecting HDIs to the real world, with iterative data collection.]**

Suppose an election is approaching, and you are interested in knowing whether the general population prefers candidate A or candidate B. There is a just-published poll in the newspaper, which states that of 100 randomly sampled people, 58 preferred candidate A and the remainder preferred candidate B.

(A) Suppose that before the newspaper poll, your prior belief was a uniform distribution. What is the 95% HDI on your beliefs after learning of the newspaper poll results?

(B) You want to conduct a follow-up poll to narrow down your estimate of the population’s preference. In your follow-up poll, you randomly sample 100 other people and find that 57 prefer candidate A and the remainder prefer candidate B. Assuming that peoples’ opinions have not changed between polls, what is the 95% HDI on the posterior?

## Solution:

Source utilities:

```r
source("../DBDA2Eprograms/DBDA2E-utilities.R")
```

Source `BernBeta.R`:

```r
source("../DBDA2Eprograms/BernBeta.R")
```

#### A.

Create a uniform prior:

```r
Prior = c(1, 1)
```

Create data vector:

```r
Data = c(rep(1, 58), rep(0, 100 - 58))
```

Find the posterior:

```r
posterior = BernBeta(Prior, Data, showCentTend = "Mode", showHDI = TRUE)
```

The HDI lies between 0.483 and 0.673 with a mode of 0.58.

#### B.

The follow-up poll contains the following results:

```r
Data2 = c(rep(1, 57), rep(0, 100 - 57))
```

Finding a new posterior:

```r
posterior2 = BernBeta(posterior, Data2, showCentTend = "Mode", showHDI = TRUE)
```

The HDI lies between 0.506 and 0.642 with a mode of 0.575.
