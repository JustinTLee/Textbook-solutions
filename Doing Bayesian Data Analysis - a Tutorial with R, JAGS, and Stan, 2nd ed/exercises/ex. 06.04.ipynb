## Problem:

**[Purpose: To explore an unusual prior and learn about the beta distribution in the process.]**

Suppose we have a coin that we know comes from a magic-trick store, and therefore we believe that the coin is strongly biased either usually to come up heads or usually to come up tails, but we donâ€™t know which. Express this belief as a beta prior. (Hint: See Figure 6.1, upper-left panel.) Now we flip the coin 5 times and it comes up heads in 4 of the 5 flips. What is the posterior distribution? (Use the R function of Section 6.6 ( BernBeta.R) to see graphs of the prior and posterior.)


## Solution:

Source utilities:

```r
source("../DBDA2Eprograms/DBDA2E-utilities.R")
```

Source `BernBeta.R`:

```r
source("../DBDA2Eprograms/BernBeta.R")
```

Generate a prior that puts emphasis towards both lower and higher probabilities (away from chance):

```r
prior = c(0.2, 0.2)
```

Create the data:

```r
data = c(rep(0, 1), rep(1, 4))
```

Generate the posterior:

```r
posterior = BernBeta(prior, data, showCentTend = "Mode", showHDI = TRUE)
```

While the HDI still includes chance (0.5), much of the interval is spread across higher probabilities with a mode of 0.941. This gives strong evidence that the coin is biased towards Heads.
