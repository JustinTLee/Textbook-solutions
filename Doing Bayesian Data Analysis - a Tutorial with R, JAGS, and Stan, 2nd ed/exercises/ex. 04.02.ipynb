## Problem:

**[Purpose: To give you some experience with random number generation in R.]**

Modify the coin flipping program in Section 4.5 RunningProportion.R to simulate a biased coin that has p(H) = 0.8. Change the height of the reference line in the plot to match p(H). Comment your code. Hint: Read the help for the sample command.

## Solution:

```r
N = 1000
pHeads = 0.8

flipSequence = sample(x = c(0, 1), prob = c(1 - pHeads, pHeads), size = N, replace = TRUE)

r = cumsum(flipSequence)
n = 1:N
runProp = r / n

plot(n, runProp, type = "o", col = "skyblue", xlim = c(1, N), ylim = c(0, 1), log = "x",
     xlab = "Flip Number",
     ylab = "Proportion Heads",
     main = "Running Proportion of Heads",
     cex.axis = 1.5, cex.lab = 1.5, cex.main = 1.5)

abline(h = pHeads, lty = "dotted")

flipLetters = paste(c("T", "H")[flipSequence[1:10] + 1], collapse = "")
displayString = paste0("Flip Sequence = ", flipLetters, "...")
text(N, .3, displayString, adj = c(1, 0.5), cex = 1.3)
text(N, .2, paste("End Proportion =", runProp[N]), adj = c(1, 0.5), cex = 1.3)
```
