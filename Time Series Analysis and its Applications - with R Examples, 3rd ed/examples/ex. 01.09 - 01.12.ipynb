```r
library(astsa)
```

#### ex. 1.9 - Moving Averages

Create white noise time series:

```r
w = rnorm(500, 0, 1)
```

Filter white noise with moving average of $v_t = \frac{1}{3}(w_{t - 1} + w_t + w_{t + 1})$:

```r
v = filter(w, sides = 2, rep(1/3, 3))
```

Plot data:

```r
par(mfrow = c(2, 1))
plot.ts(w, main = "white noise")
plot.ts(v, main = "moving average")
```

#### ex. 1.10 - Autoregressions

Create white noise time series:

```r
w = rnorm(500, 0, 1)
```

Create a recursive filter (an autoregressive equation) of the form $x_t = x_{t - 1} - 0.9x_{t - 2} + w_t$:

```r
x = filter(w, filter = c(1, -0.9), method = "recursive")[-(1:50)]
```

Plot data:

```r
plot.ts(x, main = "autoregression")
```

#### ex. 1.11 - Random Walk with Drift

Create white noise time series:

```r
w = rnorm(200, 0, 1)
wd = w + 0.2 # adding drift
```

Perform a cumulative sum $\sum_{j = 1}^{t} w_j$:

```r
x = cumsum(w)
xd = cumsum(wd) # sums drift component
```

Plot data:

```r
plot.ts(xd, ylim = c(-5, 55), main = "random walk")
lines(x)
lines(0.2*(1:200), lty = "dashed")
```

The top jagged time series is the time series with a drift component, whereas the bottom one is the one without a drift component.

#### ex. 1.12 - Signal in Noise

Create white noise time series:

```r
w = rnorm(500, 0, 1)
```

Create sinusoid:

```r
cs = 2*cos(2*pi*(1:500)/500 + 0.6*pi)
```

Plot data:

```r
par(mfrow = c(3, 1), mar = c(3, 2, 2, 1), cex.main = 1.5)
plot.ts(cs, main = expression(2*cos(2*pi*t/50 + 0.6*pi)))
plot.ts(cs + w, main = expression(2*cos(2*pi*t/50 + 0.6*pi) + N(0, 1)))
plot.ts(cs + 5*w, main = expression(2*cos(2*pi*t/50 + 0.6*pi) + N(0, 25)))
```
