## Modeling data using Kalman filter

Load the library:

```r
library(astsa);
```

Set random number generator seed for reproducibility:

```r
set.seed(1);
```

Create local level data:

```r
num = 50;
# generate time vector
Time = 1:num;

# generate two vectors of Gaussian draws
w = rnorm(num + 1, 0, 1);
v = rnorm(num, 0, 1);

# create a random walk by summing all of the Gaussian draws from w for each timepoint
mu = cumsum(w);

# add the second vector in as noise
y = mu[-1] + v;
```

Perform the Kalman filtering using `Ksmooth0` from the ASTSA library:

```r
ks = Ksmooth0(num, y, A = 1, mu0 = 0, Sigma0 = 1, Phi = 1, cQ = 1, cR = 1);
```

Plot the results:

```r
# generate subplots
par(mfrow = c(3, 1));

# generate one-step prediction plots (P(t | t - 1))
plot(Time, mu[-1], main = 'Predict', ylim = c(-5, 10));
lines(ks$xp);
lines(ks$xp + 2*sqrt(ks$Pp), lty = 2, col = 4);
lines(ks$xp - 2*sqrt(ks$Pp), lty = 2, col = 4);

# generate filtered data (P(t | t, t - 1, t - 2, ..., 1))
plot(Time, mu[-1], main = 'Filter', ylim = c(-5, 10));
lines(ks$xf);
lines(ks$xf + 2*sqrt(ks$Pf), lty = 2, col = 4);
lines(ks$xf - 2*sqrt(ks$Pf), lty = 2, col = 4);

# generate smoothed data (P(t | n, n - 1, ..., t, t - 1, ..., 1))
plot(Time, mu[-1], main = 'Smooth', ylim = c(-5, 10));
lines(ks$xs);
lines(ks$xs + 2*sqrt(ks$Ps), lty = 2, col = 4);
lines(ks$xs - 2*sqrt(ks$Ps), lty = 2, col = 4);
```
