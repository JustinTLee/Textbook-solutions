Load ASTSA package:

```r
library(astsa)
```

Plot data:

```r
plot(jj, type = "o", ylab = "Quartery Earnings per Share")
```

#### ex. 1.5 - El Nio and Fish Population

Plot data:

```r
par(mfrow = c(2, 1))
plot(soi, ylab = "", xlab = "", main = "Southern Oscillation Index (SOI)")
plot(rec, ylab = "", xlab = "", main = "Recruitment")
```

#### ex. 1.6 - fMRI Imaging

Plot data:

```r
par(mfrow = c(2, 1), mar = c(3, 2, 1, 0) + 0.5, mgp = c(1.6, 0.6, 0))
ts.plot(fmri1[, 2:5], lty = c(1, 2, 4, 5), ylab = "BOLD", xlab = "", main = "Cortex")
ts.plot(fmri1[, 6:9], lty = c(1, 2, 4, 5), ylab = "BOLD", xlab = "", main = "Thalamus & Cerebellum")
mtext("Time (1 pt = 2 sec)", side = 1, line = 2)
```
