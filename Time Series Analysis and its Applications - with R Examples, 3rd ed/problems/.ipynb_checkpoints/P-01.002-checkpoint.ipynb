## Problem:

Consider a signal-plus-noise model of the general form $x[t] = s[t] + w[t]$, where $w[t]$ is Gaussian white noise with $σ_{w}^{2} = 1$. Simulate and plot n = 200 observations from each of the following two models:

(a) $x[t] = s[t] + w[t]$, for $t = 1, \ldots , 200$, where

\begin{equation*}
s[t] = 
  \begin{cases}
    0                                                  & t = 1, \ldots, 100\\
    10 e^{-\frac{t - 100}{20}} \cos \left( \frac{2 \pi t}{4} \right) & t = 101 \ldots, 200
  \end{cases}
\end{equation*}

(b) $x[t] = s[t] + w[t]$, for $t = 1, \ldots , 200$, where

\begin{equation*}
s[t] = 
  \begin{cases}
    0                                                  & t = 1, \ldots, 100\\
    10 e^{-\frac{t - 100}{200}} \cos \left( \frac{2 \pi t}{4} \right) & t = 101 \ldots, 200
  \end{cases}
\end{equation*}

(c) Compare the general appearance of the series (a) and (b) with the earthquake series and the explosion series shown in Fig. 1.7. In addition, plot (or sketch) and compare the signal modulators (a) $e^{−t/20}$ and (b) $e^{−t/200}$, for $t = 1, 2, \ldots , 100$.

## Solution:

#### Load the library:

```r
library(astsa);
```

#### a.

Simulate the data:

```r
sta = c(rep(0, 100), 10 * exp( - ( (101:200) - 100)/(20)) * cos(2 * pi * (101:200) / 4));
wta = rnorm(200);

xta = sta + wta
```

Plot the data:

```r
plot.ts(xta, main = "signal from (a)");
```

#### a.

Simulate the data:

```r
stb = c(rep(0, 100), 10 * exp( - ( (101:200) - 100)/(200)) * cos(2 * pi * (101:200) / 4));
wtb = rnorm(200);

xtb = stb + wtb
```

Plot the data:

```r
plot.ts(xtb, main = "signal from (b)");
```

#### c.

The signals asymptotically approach $x[t] = 0$ more quickly in (a) than in (b). The dampening effect of the oscillations is greater in (a) than (b) due to the shorter time constant $\tau = 20$ in (a). An illustration of how the exponentials modify a signal are as follows:

```r
plot(exp(-(1:100)/20))
points(exp(-(1:100)/200), col = 2)
legend(78, 0.96, legend = c("exp(-t/20)", "exp(-t/200)"))
```

It is possible that this effect mimics the earthquake explosion seen in fig. 1.7.
