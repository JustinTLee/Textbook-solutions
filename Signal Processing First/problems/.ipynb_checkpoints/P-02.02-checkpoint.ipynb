## Problem:

Define $x(t)$ as

\begin{equation}
x(t) = 3 \cos(\omega_0 t - \pi/4)
\end{equation}

For $\omega_0 = \pi/5$, make a plot of $x(t)$ that is valid over the range $-10 \le t \le 20$.

## Solution:

```matlab
t = -10:0.01:20;
x = 3*cos(pi/5*t - pi/4);
plot(t, x);
title("x(t) = 3cos(\pi/5*t - \pi/4)");
xlabel("t");
ylabel("x(t)");
```
