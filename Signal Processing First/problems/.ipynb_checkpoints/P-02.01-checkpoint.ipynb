## Problem:

Figure P-2.2 is a plot of a sinusoidal wave. From the plot, determine values for the amplitude ($A$), phase ($\phi$), and frequency ($\omega_0$) needed in the representation:

\begin{equation}
x(t) = A \cos(\omega_0 t + \phi)
\end{equation}

Give the answer as numerical values, *including the units* where applicable.

## Solution:

```matlab
A = 20; %can tell from graph.
omega0 = pi/12;
t = -30:0.01:40;
phi = -pi/3;
x = A*cos(omega0*t + phi);
plot(t, x);
title("x(t) = 20cos(\pi/12*t - \pi/3)");
xlabel("time t (msec)");
ylabel("x(t)");
```
