## Problem:

Make a carefully labeled sketch for each of the following functions.

(a) Sketch $\cos (\theta)$ for values of $\theta$ in the range $0 \le \theta \le 6 \pi$.

(b) Sketch $\cos(0.2 \pi t)$ for values of $t$ such that three periods of the function are shown.

(c) Sketch $\cos(2 \pi t/T_{0})$ for values of $t$ such that three periods of the function are shown. Label the horizontal axis in terms of the parameter $T_{0}$.

(d) Sketch $\cos(2 \pi t/T_{0} + \pi/2)$ for values of $t$ such that three periods of the function are shown.

## Solution:

#### a.

```matlab
% create theta;
theta = 0:0.01:(6*pi);

%plot
plot(theta, cos(theta), 'k');
title("cos(\theta) from \theta = 0 to \theta = 6\pi");
xlabel("\theta");
ylabel("cos(\theta)");
xticks([0 pi 2*pi 3*pi 4*pi 5*pi 6*pi]);
xticklabels({'0', '\pi', '2\pi', '3\pi', '4\pi', '5\pi', '6\pi'});
xlim([0 6*pi]);
ylim([-1 1]);
```

#### b.-c.

Find the frequency:<br/>
\begin{align*}
\omega &= 2 \pi f \\ 
0.2 \pi &= 2 \pi f \\
0.1 \: \text{Hz} &= f \\
\end{align*}

Find the period:<br>
\begin{align*}
T &= \frac{1}{f} \\
T &= \frac{1}{0.1 \: \text{Hz}} \\
T &= 10 \: \text{s} \\
\end{align*}

```matlab
% t and cos(0.2*pi*t)
T = 10; % period of 10 seconds
t = 0:0.1:(3*T);
xt = cos(0.2*pi*t);

% plot
plot(t, xt, 'k');
title("cos(0.2\pi t) for 3 periods");
xlabel("t (sec)");
ylabel("cos(0.2\pi t)");
xticks([0 10 20 30]);
ylim([-1 1]);
```

#### d.

```matlab
% cos(0.2*pi*t + pi/2)
xt = cos(0.2*pi*t + pi/2);

% plot
plot(t, xt, 'k');
title("cos(0.2\pi t + \pi/2) for 3 periods");
xlabel("t (sec)");
ylabel("cos(0.2\pi t + \pi/2)");
xticks([0 10 20 30]);
ylim([-1 1]);
```
