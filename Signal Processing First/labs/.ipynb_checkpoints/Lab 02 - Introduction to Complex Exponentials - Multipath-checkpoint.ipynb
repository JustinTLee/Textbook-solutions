### 3.1. Complex Numbers

Use $z_{1} = 10e^{-j2\pi/3}$ and $z_{2} = -5 + j5$ for all parts of this section.

#### a. Enter the complex numbers $z_{1}$ and $z_{2}$ in `MATLAB`. Plot them with `zvect()` and print them with `zprint()`. Whenever you make a plot with `zvect()` or `zcat()`, it is helpful to provide axes for reference. An x-y axis and the unit circle can be superimposed on your `zvect()` plot by doing the following: `hold on, zcoords, ucplot, hold off`

````matlab
z1 = 10*exp(-j*2/3*pi);
z2 = -5 + j*5;

z = [z1, z2];

figure(1);
zvect(z, 'r');
hold on;
ucplot(1, '--k');
zcoords('k');
hold off;
title("Plot of z_{1} and z_{2}");
xlabel("real");
ylabel("complex");

zprint(z);
````

#### b. The function `zcat()` can be used to plot vectors in a “head-to-tail” format. Execute the statement `zcat([j,-1,-2j,1]);` to see how `zcat()` works when its input is a vector of complex numbers.

````matlab
figure(2);
zcat([j, -1, -2*j, 1], 'r');
title("zcat() function");
xlabel("Real");
ylabel("Complex");
````

#### c. Compute $z_{1} + z_{2}$ and plot the sum using `zvect()`. Then use `zcat()` to plot $z_{1}$ and $z_{2}$ as 2 vectors head-to-tail, thus illustrating the vector sum. Use hold on to put all 3 vectors on the same plot. If you want to see the numerical value of the sum, use `zprint()` to display it.

````matlab
z_sum = sum(z);
figure(3);
zvect(z_sum, 'r');
hold on;
zcat(z);
hold off;
title("Sum of complex vectors");
xlabel("Real");
ylabel("Complex");

zprint(z_sum);
````

#### d. Compute $z_{1}z_{2}$ and plot the answer using `zvect()` to show how the angles of $z_{1}$ and $z_{2}$ determine the angle of the product. Use `zprint()` to display the result numerically.

````matlab
z_prod = z1*z2;
figure(4);
zvect(z_prod, 'r');
title("Product of complex vectors");
xlabel("Real");
ylabel("Complex");

zprint(z_prod);
````

#### e. Compute $z_{2}/z_{1}$ and plot the answers using `zvect()` to show how the angles of $z_{1}$ and $z_{2}$ determine the angle of the quotient. Use `zprint()` to display the result numerically.

````matlab
z_quot = z1./z2;
figure(5);
zvect(z_quot, 'r');
title("Quotient of complex vectors");
xlabel("Real");
ylabel("Complex");

zprint(z_quot);
````

#### f. Compute the conjugate $z^{∗}$ for both $z_{1}$ and $z_{2}$ and plot the results. In `MATLAB`, see `help conj`. Display the results numerically with `zprint`.

````matlab
z1_conj = conj(z1);
z2_conj = conj(z2);

zprint(z1_conj);
zprint(z2_conj);
````

#### g. Compute the inverse $1/z$ for both $z_{1}$ and $z_{2}$ and plot the results. Display the results numerically with `zprint`.

````matlab
z_inv = [1/z1, 1/z2];

figure(6);
zvect(z_inv, 'r');
title("Inverse of complex vectors");
xlabel("Real");
ylabel("Complex");

zprint(z_inv);
````

#### h. Make a 2 × 2 subplot that displays the following four plots in one figure window: (i) $z_{1}$ and $z_{2}$ (ii) $z_{1}^{*}$ and $z_{2}^{*}$ on the same plot; (iii) $1/z_{1}$ and $1/z_{2}$ on the same plot; and (iv) $z_{1}z_{2}$. Add a unit circle and x-y axis to each plot for reference.

````matlab
figure(7);

subplot(2, 2, 1);
zvect(z, 'r');
hold on;
ucplot(1, '--k');
zcoords('k');
hold off;
title("i");
xlabel("real");
ylabel("complex");
subplot(2, 2, 2);
zvect([z1_conj, z2_conj], 'r');
hold on;
ucplot(1, '--k');
zcoords('k');
hold off;
title("ii");
xlabel("real");
ylabel("complex");
subplot(2, 2, 3);
zvect(z_inv, 'r');
hold on;
ucplot(1, '--k');
zcoords('k');
hold off;
title("iii");
xlabel("Real");
ylabel("Complex");
subplot(2, 2, 4);
zvect(z_prod, 'r');
hold on;
ucplot(1, '--k');
zcoords('k');
hold off;
title("iv");
xlabel("Real");
ylabel("Complex");
````

### 3.3. Vectorization

The following code is presented in the Pre-Lab:

````matlab
y = cos(0.001*pi*(1:200).^2);

figure(8);
plot(1:200, y, 'ro-');
title("Vectorization of cos(0.001 pi x^2)");
xlabel("x");
ylabel("y");
````

The Pre-Lab gives the following code:

```
%--- make a plot of a weird signal
N = 200;
for k=1:N
    xk(k) = k/50;
    rk(k) = sqrt( xk(k)*xk(k) + 2.25 );
    sig(k) = exp(j*2*pi*rk(k));
end
plot( xk, real(sig), ’mo-’ )
```

It can be vectorized in the following manner:

````matlab
rk = sqrt(((1:200)/50).^2 + 2.25);
sig = exp(j*2*pi*rk.^2);

plot((1:200)/50, real(sig), 'mo-');
title("Exponential of function");
xlabel("x");
ylabel("r(x)");
````
